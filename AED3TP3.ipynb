```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t1 = pd.read_csv("./salida/clusteringSWEEP-SET-Av2.csv", index_col="n")
t1["diferencia-costo"] = t1["costo-sol"] - t1["costo-opt"]
t1["diferencia-k"] = t1["k-sol"] - t1["k-opt"]
t1
#listaPlot = [i for i in range()]
#xticks=listaPlot,
```

```python
graf1 = t1[["costo-sol", "costo-opt"]].plot(figsize=(10, 5), grid=True, title="Comparacin de los costos del algoritmo sweep con el costo optimo del SET A")
graf1.set_xlabel("Cantidad de clientes")
graf1.set_ylabel("Costo de la solucin")
graf1.legend(["Costo Algortimo sweep","Costo Optimo"])
plt.savefig("COMPARACION-SWEEP-SET-A.pdf")
```
